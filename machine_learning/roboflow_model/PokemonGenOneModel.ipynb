```python
model = YOLO('yolov8n-cls.pt')
```

```python
save_dir = './train'
os.makedirs(save_dir, exist_ok=True)

model.train(
    data='./Pokemon-1',
    epochs=1,
    imgsz=224,
    device='cpu',
    batch=16,
    patience=15,
    project=f'{HOME}',
    save_dir=save_dir
)
```

```python
model.predict(
    source='./Pokemon-1/test/Psyduck',
    imgsz=1280,
    device='cpu'
)
```

```python
model.val(
    data='./Pokemon-1',
    device='cpu'
)
```

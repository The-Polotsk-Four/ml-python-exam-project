```python
model = YOLO('yolo26m-cls.pt')
```

```python
save_dir = './model_training'
model = YOLO(f'{save_dir}/weights/last.pt')  # load last checkpoint
os.makedirs(save_dir, exist_ok=True)

model.train(
    data='./Pokemon-1',
    epochs=10,
    imgsz=640,
    device='cuda',
    batch=8,
    project=f'{HOME}',
    save=True,
    save_dir=save_dir,
    time=10
)
```

```python
model.predict(
    source='./Pokemon-1/test/Dewgong',
    imgsz=1280,
    device='cuda'
)
```

```python
model.val(
    data='./Pokemon-1',
    device='cuda'
)
```
